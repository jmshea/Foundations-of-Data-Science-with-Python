## Proofs Omitted in Text

Proof that all events $B_j, ~B_k,~j \ne k$ are mutually exclusive
$$
B_j \cap B_k = (B \cap A_j) \cap (B \cap A_k) = B \cap (A_j \cap A_k) = B \cap \emptyset =\emptyset.
$$



Proof that the union of all events $B_j$ is equal to $B$
$$
\bigcup_j B_j = \bigcup_j \left(A_j \cap B\right )  = B \cap \left( \bigcup_j A_j \right) = B \cap S = B.
$$
